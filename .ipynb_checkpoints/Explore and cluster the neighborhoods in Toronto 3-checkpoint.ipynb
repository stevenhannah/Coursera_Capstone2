```python
import pandas as pd
```

```python
df_final=pd.read_csv('df_final.csv')
df_geo=pd.read_csv('https://cocl.us/Geospatial_data')
```

```python
df_final=df_final.iloc[:,1:]
```

```python
df_geo.head()
```

Merge the dataframe

```python
df_finalwithgeo=df_final.join(df_geo.set_index('Postal Code'),on='Postcode')
```

```python
df_finalwithgeo
```

```python
import folium

m = folium.Map(
    location=[43.753259,-79.329656],
    zoom_start=11,
)
```

Use for loop to set the marker 

```python
for i in range(len(df_finalwithgeo.index)):
    html=df_finalwithgeo.Postcode[i]
    folium.Marker(
        [df_finalwithgeo.Latitude[i], df_finalwithgeo.Longitude[i]],
        popup=folium.Popup(html,max_width=450)
    ).add_to(m)
    
m.add_child(folium.ClickForMarker(popup='Waypoint'))    
```
